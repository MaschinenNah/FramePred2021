```python
def load_and_save(id, filename):
  text = load_etext(id)
  text = strip_headers(text)
  text = text.strip()
  with open("raw/" + filename + ".txt", 'w') as f:
    f.write(text)
```

```python
def crop_file(path, from_start, from_end=False):
  with open(path) as f:
    lines = f.readlines()
  
  if from_end:
    lines = lines[from_start:from_end-1]
    text = "".join(lines)[:-1]
  else:
    lines = lines[from_start:]
    text = "".join(lines)

  with open(path, 'w') as f:
    f.write(text)
```
